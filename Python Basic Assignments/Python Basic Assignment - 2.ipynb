## Q10. How can you tell the difference between break and continue?

# Ans:-
    The primary difference between break and continue statement in is that the break statement leads to an immediate exit of the innermost switch or enclosing loop. On the other hand, the continue statement begins the next iteration of the while, enclosing for, or do loop.

```python
# use of break
for i in range(10):
    if(i==5):
        break    # The break statement will move the execution outside and just after a loop.
    print(i)
    
print('Breaked')

#use of  continue   
for i in range(10):
    if(i==7):  # 7 value is skipped 
        continue   # # The continue statement will move the execution to the start of the loop.
    print(i)
```

## Q11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

#### Ans:-
     The output we can conclude that they all do the same thing. The range(10) call ranges from 0 up to (but not including) 10, range(0, 10) tells the loop to start at 0, and range(0, 10, 1) tells the loop to increase the variable by 1 on each iteration.

```python
for i in range(10):
    print(i)
print("*********") 
for i in range(0,10):
    print(i)
print("*********")
for i in range(0,10,1):
    print(i)
```
